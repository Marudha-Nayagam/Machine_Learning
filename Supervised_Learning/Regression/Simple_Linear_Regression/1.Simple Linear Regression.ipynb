```python
# Step 1: Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the Dataset
data = pd.read_csv("Salary_Data.csv")
```

```python
data
```

```python
# Independent and Dependent Variables
indep = data[['YearsExperience']] # independent variable
print(indep.head())
dep = data[['Salary']] # dependent variable
print(dep.head())
```

```python
# Step 3: Data Visualization
# Scatter Plot: to display the relationship between two different numerical values
plt.scatter(indep, dep, color='red')
plt.xlabel("Years of Experience", fontsize=14)
plt.ylabel("Salary", fontsize=14)
plt.title("Years of Experience vs Salary")
plt.show()
```

```python
# Step 4: Train and Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(indep, dep, test_size=1/3, random_state=0)
```

```python
# print(data.shape)
# print("Length of x train: ",len(x_train))
# print("Length of x test: ",len(x_test))
```

```python
# Step 5: Model Training
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train) # trianing the model
```

```python
# Step 6: Model Parameters
weight = model.coef_
bias = model.intercept_

print(f"Weight (Slope) = {weight}")
print(f"Bias (Intercept) = {bias}")
```

```python
# Step 7: Predictions
y_pred = model.predict(x_test)
y_pred
```

```python
# Step 8: Model Evaluation
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"R² Score = {r2}")
```

```python
# Step 9: Visualizing Regression Line
plt.scatter(x_train, y_train)
plt.plot(x_train, model.predict(x_train))
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Simple Linear Regression Line")
plt.show()
```

```python
# Step 10: Save the Model
import pickle
filename = "salary_prediction_model.pkl"
pickle.dump(model, open(filename,'wb'))
```

```python
# Step 11: Load the Model
loaded_model = pickle.load(open(filename,"rb"))
```

```python
# Step 12: User Input Prediction
experience = float(input("Enter Years of Experience: "))
future_prediction = loaded_model.predict([[experience]])

print(f"Predicted Salary = {future_prediction[0][0]}")
```
