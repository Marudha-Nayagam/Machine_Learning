```python
import pickle
filename = 'profit_prediction_model_mlr.pkl'
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
# Get User Input
rd_input = float(input("R&D: "))
adminis_input = float(input("Administration: "))
marketing_input = float(input("Marketing: "))
florida = int(input("State_Florida 0 / 1: "))
newyork = int(input("State Newyork 0 / 1: "))
```

```python
profit = loaded_model.predict([[rd_input, adminis_input, marketing_input, florida, newyork]])
```

```python
print("Profit Prediction: ", profit)
```
